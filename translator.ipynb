```python
import re
```

```python
def translate_Arabic(text):
    buckwalterMod = {
        'ء': '2', 'ا': 'a', 'إ': 'a',
        'أ': 'a', 'آ': 'a', 'ب': 'b',
        'ة': 't', 'ت': 't', 'ث': 't',
        'ج': 'j', 'ح': '7', 'خ': 'kh',
        'د': 'd', 'ذ': 'd', 'ر': 'r',
        'ز': 'z', 'س': 's', 'ش': 'ch',
        'ص': 'S', 'ض': 'd', 'ط': 'T',
        'ظ': 'd', 'ع': '3', 'غ': 'gh',
        'ف': 'f', 'ق': '9', 'ك': 'k',
        'ل': 'l', 'م': 'm', 'ن': 'n',
        'ه': 'h', 'ؤ': '2', 'و': 'ou',
        'ى': 'y', 'ئ': 'c', 'ي': 'y',
        }
    for k, v in buckwalterMod.items():
        text = re.sub(k, v, text)
    return(text)
```

```python
dictionnaire_darija = pd.read_csv('dictionnaire_darija.csv')
dictionnaire_darija
```

```python
translation2 = dictionnaire_darija["eng"].loc[dictionnaire_darija['darija'] == "dght"]
translation2
```

```python
def translate(sentence):
    splited_sentence = sentence.split()
    translated_sentence = ""
    for word in splited_sentence:
        try:
            translation = dictionnaire_darija["eng"].loc[dictionnaire_darija['darija'] == word].item()
            translated_sentence += translation + " "
            #translated_sentence = translated_sentence + translation + " "
        except Exception:
            pass #for words that dont figure in our csv file
    return translated_sentence
```

### converting dataset from arabic letters to latin letters

```python
darija_dataset = pd.read_csv('Sentiment_Anaysis.csv')
darija_dataset
```

```python
darija_dataset = darija_dataset.drop(darija_dataset.columns[0], axis=1)
darija_dataset
```

```python
darija_dataset=darija_dataset.replace(to_replace="pos",value="1")
darija_dataset=darija_dataset.replace(to_replace="neu",value="1")
darija_dataset=darija_dataset.replace(to_replace="neg",value="0")
```

```python
darija_dataset
```

```python
sentences = darija_dataset['Twits']
```

```python
latin_sentences = []
for i in sentences:
    latin_sentences.append(translate_Arabic(i))
```

```python
latin_sentences
```

```python
latin_sentences = pd.DataFrame(latin_sentences)
latin_sentences
```

```python
darija_dataset = pd.concat([darija_dataset, latin_sentences], axis=1)
darija_dataset
```

```python
darija_dataset = darija_dataset.drop(darija_dataset.columns[0], axis=1)
darija_dataset
```

```python
darija_dataset = darija_dataset.rename(columns={0: 'sentence'})
darija_dataset
```

```python
columns_titles = ["sentence", "label"]
darija_dataset=darija_dataset.reindex(columns=columns_titles)
darija_dataset
```

```python
darija_dataset.head(50)
```

```python
sentences = darija_dataset["sentence"]
sentences
```

```python
english_sentences = []
for i in sentences:
    english_sentences.append(translate(i))
```

```python
english_sentences
```

```python
english_sentences = pd.DataFrame(english_sentences)
english_sentences = english_sentences.rename(columns={0: 'english_translation'})
english_sentences
```

```python
final_dataset = pd.concat([english_sentences, darija_dataset], axis=1)
final_dataset
```

```python
final_dataset = final_dataset.drop(['sentence'],axis=1)
final_dataset
```

```python
final_dataset.to_csv('final_datadet.csv')
```
